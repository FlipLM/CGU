```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/FlipLM/CGU/main/dados.csv'
df = pd.read_csv(url, sep=';')
df = df.dropna(subset=['Pedido'])
```

# **Exercício 1**
Pedidos de acesso à informação:





```python
# @title Qual é o número total de pedidos?

numero_total_pedidos = len(df)
numero_total_pedidos
```

```python
# @title Criar uma coluna em memória com o tamanho do texto de cada pedido
pedidos_existentes = 0
tamanhos_pedidos = []

for texto in df['Pedido']:
    if isinstance(texto, str):
        pedidos_existentes += 1
        tamanho_pedido = len(texto)
        tamanhos_pedidos.append(tamanho_pedido)
    else:
        tamanhos_pedidos.append(None)

df['Tamanho do Pedido'] = tamanhos_pedidos

df
```

```python
# @title Qual o pedido mais extenso e qual o seu tamanho?
pedido_mais_extenso = ''
tamanho_pedido_mais_extenso = 0

for texto in df['Pedido']:
    if isinstance(texto, str) and len(texto) > tamanho_pedido_mais_extenso:
        tamanho_pedido_mais_extenso = len(texto)
        pedido_mais_extenso = texto

pedido_mais_extenso
```

# **Exercício 2**

> Tempo de Resposta:



```python
# @title Qual é o maior tempo de resposta?
maior_tempo_resposta = df['Dias para Resposta'][0]
for tempo in df['Dias para Resposta']:
    if tempo > maior_tempo_resposta:
        maior_tempo_resposta = tempo

maior_tempo_resposta
```

```python
# @title Qual é o menor tempo de resposta?
menor_tempo_resposta = df['Dias para Resposta'][0]
for tempo in df['Dias para Resposta']:
    if tempo < menor_tempo_resposta:
        menor_tempo_resposta = tempo
menor_tempo_resposta
```

```python
# @title Qual é o tempo médio de resposta?
soma_tempo_resposta = 0
for tempo in df['Dias para Resposta']:
    soma_tempo_resposta += tempo

media_tempo_resposta = soma_tempo_resposta / len(df['Dias para Resposta'])

media_tempo_resposta
```

```python
# @title Qual a moda do tempo médio de resposta?
contagem_tempo = {}
for tempo in df['Dias para Resposta']:
    if tempo in contagem_tempo:
        contagem_tempo[tempo] += 1
    else:
        contagem_tempo[tempo] = 1

moda_tempo_resposta = None
max_contagem = 0
for tempo, contagem in contagem_tempo.items():
    if contagem > max_contagem:
        moda_tempo_resposta = tempo
        max_contagem = contagem

moda_tempo_resposta
```

```python
# @title Qual a mediana do tempo médio de resposta?
tempos_ordenados = sorted(df['Dias para Resposta'])
tamanho_lista = len(tempos_ordenados)
if tamanho_lista % 2 == 0:
    mediana = (tempos_ordenados[tamanho_lista // 2 - 1] + tempos_ordenados[tamanho_lista // 2]) / 2
else:
    mediana = tempos_ordenados[tamanho_lista // 2]

mediana
```

# **Exercício 3**

> Identidade de Gênero. (instruções: não é permitido o uso da função unique ou similar)



```python
# @title Quais identidades de gênero estão presentes nos dados?
identidades_genero = []

for identidade in df['Identidade de gênero']:
    if identidade not in identidades_genero:
        identidades_genero.append(identidade)

identidades_genero
```

```python
# @title Exiba a porcentagem de cada identidade de gênero em relação ao total de pedidos.
total_pedidos = len(df)

porcentagens = {}
for identidade in identidades_genero:
    contagem = 0
    for valor in df['Identidade de gênero']:
        if valor == identidade:
            contagem += 1
    porcentagem = (contagem / total_pedidos) * 100
    porcentagens[identidade] = porcentagem

porcentagens
```

# **Exercício 4**

> Ranking.



```python
# @title Liste os 5 Estados com maior número de pedidos de acesso à informação, juntamente com a quantidade correspondente, em ordem decrescente

estados_contagem = {}
for estado in df['Estado ']:
    if estado in estados_contagem:
        estados_contagem[estado] += 1
    else:
        estados_contagem[estado] = 1

estados_ranking = sorted(estados_contagem.items(), key=lambda x: x[1], reverse=True)


for i, (estado, contagem) in enumerate(estados_ranking[:5], 1):
    print(f"{i}. {estado}: {contagem} pedidos")
```

```python
# @title Quais cidadãos possuem o maior número de pedidos de acesso à informação, liste os 3 mais frequentes e a quantidade de pedidos de cada

cidadaos_contagem = {}
for cidadao in df['Nome do Cidadão']:
    if cidadao in cidadaos_contagem:
        cidadaos_contagem[cidadao] += 1
    else:
        cidadaos_contagem[cidadao] = 1

cidadaos_ranking = sorted(cidadaos_contagem.items(), key=lambda x: x[1], reverse=True)


for i, (cidadao, contagem) in enumerate(cidadaos_ranking[:3], 1):
    print(f"{i}. {cidadao}: {contagem} pedidos")
```
